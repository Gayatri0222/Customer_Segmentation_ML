This step applies log transformation to highly skewed numerical features in order to reduce skewness and stabilize variance.

```python
# Function to cap outliers
def cap_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return np.clip(col, lower, upper)
```

```python
# Features where outliers matter (distance-based model)
features = ['Age','Income', 'Total_Products', 'Total_Purchases', 'Recency']

# Apply capping
for col in features:
    df1[col] = cap_outliers(df1[col])
```

By capping outliers in Income, Total_Products, Total_Purchases, and Recency, we ensure that the clustering process is more stable, balanced, and representative of true customer behavior.

### **Model Building**

```python
# Selecting only the important features for customer segmentation
# These features directly represent customer behavior and value
clustering_features = ['Age', 'Total_Products', 'Total_Purchases', 'Income', 'Recency']

# Creating feature matrix X using selected columns
X = df1[clustering_features]

# Scaling the data
# KMeans, Hierarchical, and DBSCAN are distance-based algorithms,
# so scaling is required to bring all features to the same scale
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

In this step, we selected key numerical features that represent customer demographics
and purchasing behavior. These features include Age, Income, Recency, Total Products,
and Total Purchases.

Since clustering algorithms such as KMeans, Hierarchical, and DBSCAN are distance-based,
feature scaling is necessary. StandardScaler was used to standardize the features so that
each variable contributes equally to the clustering process.

```python
# Import KMeans algorithm
from sklearn.cluster import KMeans

# Import matplotlib for plotting
import matplotlib.pyplot as plt

# This list will store WCSS values for different k
# WCSS = Within-Cluster Sum of Squares
wcss = []

# Try different number of clusters from 1 to 10
for k in range(1, 11):
    
    # Create KMeans model for current k
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model on scaled data
    # We use scaled data because KMeans is distance-based
    kmeans.fit(X_scaled)
    
    # inertia_ gives WCSS value for current k
    wcss.append(kmeans.inertia_)

# Plot the Elbow graph
plt.figure(figsize=(8, 5))

# X-axis: number of clusters (k)
# Y-axis: WCSS value
plt.plot(range(1, 11), wcss, marker='o')

# Label the axes
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')

# Title of the plot
plt.title('Elbow Method to Determine Optimal Number of Clusters')

# Show the plot
plt.show()
```

The Elbow Method is used to find the optimal number of clusters in KMeans clustering.
We calculate WCSS for different values of k.
As the number of clusters increases, WCSS decreases because clusters become tighter.
The point where the decrease in WCSS slows down and forms an elbow shape is considered the optimal number of clusters.
Based on the elbow plot, k = 3 was chosen for KMeans clustering.

```python
# Import PCA for dimensionality reduction
from sklearn.decomposition import PCA
import pandas as pd

# Apply PCA to reduce features into 2 principal components
# n_components=2 is used for visualization and clustering
pca = PCA(n_components=2, random_state=42)

# Transform the scaled data into PCA space
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for PCA output
# PC1 and PC2 are the new reduced dimensions
pca_df = pd.DataFrame(
    X_pca,
    columns=['PC1', 'PC2']
)

# Check how much information (variance) is preserved
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Explained:", pca.explained_variance_ratio_.sum())

# Visualize PCA result
plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', data=pca_df, alpha
=0.5)
plt.title("PCA Result - 2D Scatter Plot")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

Principal Component Analysis (PCA) was applied to reduce the dimensionality of the dataset.
The original five features were transformed into two principal components (PC1 and PC2)
that capture the maximum variance in the data.

Using PCA improves clustering performance by reducing noise and correlation among features
and also allows easy visualization of customer segments.

The explained variance ratio indicates the proportion of information retained by each
principal component.

### **Model 1 - Kmeans Clustering**

```python
# Import KMeans clustering algorithm
from sklearn.cluster import KMeans

# Create KMeans model with 3 clusters
# n_clusters=3 because we want to segment customers into 3 groups
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model on PCA-transformed data and predict cluster labels
kmeans_labels = kmeans.fit_predict(X_pca)

# Add the cluster labels back to the original dataframe
df1['KMeans_Cluster'] = kmeans_labels
```

### KMeans Clustering

KMeans clustering was applied to group customers into three distinct segments
based on their purchasing behavior and demographics. The clustering was performed
on PCA-transformed data to improve separation and reduce noise.

Each customer was assigned a cluster label, which was added back to the dataset
for further analysis and interpretation.


**Why KMeans Was Used**

Simple and fast clustering algorithm

Works well for large datasets

Easy to interpret results

Suitable when number of clusters is known

```python
# Import evaluation metrics for clustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Calculate evaluation metrics for KMeans clustering

# Silhouette Score:
# Measures how well each point fits within its cluster
# Higher value → better clustering
k_sil = silhouette_score(X_pca, kmeans_labels)

# Davies-Bouldin Index:
# Measures similarity between clusters
# Lower value → better clustering
k_db = davies_bouldin_score(X_pca, kmeans_labels)

# Calinski-Harabasz Score:
# Measures cluster separation and compactness
# Higher value → well-defined clusters
k_ch = calinski_harabasz_score(X_pca, kmeans_labels)

# Display KMeans evaluation scores
print("KMeans Scores")
print("Silhouette Score:", round(k_sil, 3))
print("Davies-Bouldin Index:", round(k_db, 3))
print("Calinski-Harabasz Score:", round(k_ch, 1))

```

We evaluated the quality of KMeans clustering using three metrics.

Silhouette Score checks how well customers fit in their clusters.

Davies-Bouldin Index checks how similar clusters are to each other.

Calinski-Harabasz Score checks how compact and separated clusters are.

Using multiple metrics gives a reliable evaluation instead of trusting one score.

| Metric            | Ideal Value     | Meaning           |
| ----------------- | --------------- | ----------------- |
| Silhouette        | Closer to **1** | Better clustering |
| Davies-Bouldin    | **Lower**       | Less overlap      |
| Calinski-Harabasz | **Higher**      | Clear separation  |

```python
# Import functions required to build dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage

# Create linkage matrix using Ward's method
# Ward's method minimizes variance within clusters
linked = linkage(X_scaled, method='ward')
```

Hierarchical clustering builds clusters step by step.

The linkage function calculates the distance between clusters.

We used Ward’s method because it:

Minimizes variance inside clusters

Works well with scaled numerical data

The output (linked) is used to plot a dendrogram

```python
# Plotting the dendrogram to visualize hierarchical clustering structure
plt.figure(figsize=(12, 6))

dendrogram(
    linked,                    # linkage matrix created earlier
    truncate_mode='lastp',      # show only the last merged clusters
    p=12,                       # number of final clusters/groups to display
    leaf_rotation=45,           # rotate labels for readability
    leaf_font_size=10           # font size of labels
)

# Adding title and axis labels
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Cluster Groups")
plt.ylabel("Euclidean Distance")

plt.show()
```

This dendrogram visually shows how customers are grouped step by step.

The height of the vertical lines represents the distance between clusters.

Large vertical gaps indicate natural cluster separation.

By cutting the dendrogram at an appropriate height, we can clearly justify
three major customer clusters.

This validates the choice of 3 clusters used in our model.

```python
# Import Agglomerative (Hierarchical) Clustering algorithm
from sklearn.cluster import AgglomerativeClustering

# Apply Agglomerative Clustering with 3 clusters
# Number of clusters is chosen based on dendrogram analysis
# Ward linkage minimizes variance within clusters
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')

# Fit the model on PCA-reduced data and get cluster labels
agg_labels = agg.fit_predict(X_pca)

# Store the hierarchical cluster labels in the dataframe
df1['Agg_Cluster'] = agg_labels
```

### **MODEL 3: DBSCAN**

### DBSCAN Clustering

In this section, we apply **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** to our dataset after dimensionality reduction using PCA.  

DBSCAN is a density-based clustering algorithm that groups together points that are **closely packed** and marks points that lie alone in low-density regions as **outliers** (-1 cluster label).  

It is particularly useful when the clusters are of **arbitrary shape** and we want to automatically detect **noise points**.


```python
# k for DBSCAN eps selection
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Step 1: choose k = min_samples
k = 6   # because you have 5 features → min_samples = 6

# Step 2: fit nearest neighbors
nbrs = NearestNeighbors(n_neighbors=k)
nbrs.fit(X_scaled)

# Step 3: compute k-distance
distances, _ = nbrs.kneighbors(X_scaled)

# Step 4: sort distances
k_distances = np.sort(distances[:, k-1])

# Step 5: plot k-distance graph
plt.plot(k_distances)
plt.xlabel("Sorted points")
plt.ylabel("k-distance")
plt.title("k-Distance Graph for DBSCAN eps selection")
plt.show()
```

There's a sharp bend at 0.7 so we will go with 0.7 eps

```python
# Import DBSCAN clustering algorithm
from sklearn.cluster import DBSCAN

# Import StandardScaler for feature scaling
from sklearn.preprocessing import StandardScaler

# Import clustering evaluation metrics
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score


# STEP 1: Scale the data
# DBSCAN is distance-based, so scaling is very important
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# STEP 2: Apply DBSCAN
# eps = 0.7 → we chose 0.7 because it balances cluster density and noise
# min_samples = 6 → we have 5 features, so min_samples=6 is reasonable
dbscan = DBSCAN(eps=0.7, min_samples=6)

# Fit DBSCAN and get cluster labels
# Noise points are labeled as -1
labels = dbscan.fit_predict(X_scaled)


# STEP 3: Remove noise points
# We keep only points that are NOT noise (-1)
mask = labels != -1


# STEP 4: Calculate Silhouette Score
# Measures how well points fit within their cluster
sil_score = silhouette_score(X_scaled[mask], labels[mask])


# STEP 5: Calculate Davies-Bouldin Index
# Lower value = better clustering
db_score = davies_bouldin_score(X_scaled[mask], labels[mask])


# STEP 6: Calculate Calinski-Harabasz Score
# Higher value = better clustering
ch_score = calinski_harabasz_score(X_scaled[mask], labels[mask])


# STEP 7: Print all scores
print("DBSCAN Silhouette Score:", sil_score)
print("DBSCAN Davies-Bouldin Score:", db_score)
print("DBSCAN Calinski-Harabasz Score:", ch_score)
```

### DBSCAN Cluster Results

After applying DBSCAN to our PCA-transformed data, we get the following cluster distribution:

| Cluster Label | Number of Customers | Description           |
|---------------|------------------|----------------------|
| 0             | 2235             | Main cluster of customers with similar behavior |
| -1            | 5                | Noise points (outliers) that do not belong to any cluster |

### Interpretation:

1. **Cluster 0** contains the majority of the data points. This indicates that most customers share similar purchasing patterns or behavior.  
2. **Cluster -1** represents a very small number of **outliers**—customers whose behavior is significantly different from the main group.  
3. The fact that DBSCAN identifies noise points (-1) is one of its advantages over K-Means, which forces every point into a cluster.  
4. Since there is only **one main cluster**, this suggests that our dataset is **dense and homogeneous**, with very few anomalies.


```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(7,5))

# Scatter plot of PCA components
# x = first principal component
# y = second principal component
# hue = cluster labels from DBSCAN
sns.scatterplot(
    x=X_pca[:,0],
    y=X_pca[:,1],
    hue=labels,          # Different colors for different clusters
    palette='tab10',        # Color palette
    legend='full',          # Show legend
    s=50                    # Size of scatter points
)

# Plot title and axis labels
plt.title("DBSCAN Clustering")
plt.xlabel("PCA Component 1 (PC1)")
plt.ylabel("PCA Component 2 (PC2)")
plt.show()
```

### Interpretation of the Plot:

1. Each color represents a **DBSCAN cluster**.  
2. Points labeled **-1** (outliers) will appear in a separate color.  
3. DBSCAN can detect **dense clusters** and **ignore noise points**, unlike K-Means which assigns every point to a cluster.  
4. In our dataset:
   - Most points are in a **single dense cluster** (Cluster 0).  
   - Only a few points are **outliers (-1)**, indicating unusual customer behavior.  

This visual makes it easy to **understand customer segmentation** and detect anomalies in purchasing behavior.


```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Mask to exclude noise points (-1) from evaluation
mask = labels != -1

# Check if there are at least 2 clusters (necessary for evaluation metrics)
if len(set(labels[mask])) > 1:
    # Silhouette Score: higher is better (max 1)
    d_sil = silhouette_score(X_pca[mask], labels[mask])
    
    # Davies-Bouldin Score: lower is better (min 0)
    d_db  = davies_bouldin_score(X_pca[mask], labels[mask])
    
    # Calinski-Harabasz Score: higher is better
    d_ch  = calinski_harabasz_score(X_pca[mask], labels[mask])

    # Print results
    print("DBSCAN Evaluation Scores")
    print("Silhouette Score:", round(d_sil, 3))
    print("Davies-Bouldin Score:", round(d_db, 3))
    print("Calinski-Harabasz Score:", round(d_ch, 1))

else:
    # If only one cluster exists, metrics cannot be calculated
    print("DBSCAN did not form valid clusters for evaluation")
```

### Evaluation of DBSCAN Clusters

When we ran the evaluation metrics (Silhouette, Davies-Bouldin, Calinski-Harabasz), we got the message:

**"DBSCAN did not form valid clusters for evaluation"**  

### Why this happened:

1. **DBSCAN output:**
   - Cluster 0 → 2235 points (majority of customers)  
   - Cluster -1 → 5 points (noise/outliers)

2. **Evaluation metrics requirement:**
   - Silhouette, Davies-Bouldin, and Calinski-Harabasz **require at least 2 clusters** to compute meaningful scores.  
   - In our case, DBSCAN found **only 1 main cluster**, so these metrics cannot be calculated.

3. **Interpretation:**
   - Our customer data is **dense and homogeneous**, meaning most customers behave similarly.  
   - DBSCAN’s main advantage here was **detecting outliers (5 customers)** rather than forming multiple clusters.  
   - This is **expected behavior** with DBSCAN on datasets with **very few distinct clusters**.

Therefore, KMeans with 3 clusters was selected for final customer segmentation.

```python
import joblib
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Create FULL pipeline (Scaler + PCA + KMeans)
pipeline = Pipeline(
    steps=[
        ('scaler', StandardScaler()),
        ('pca', PCA(n_components=2, random_state=42)),
        ('model', KMeans(n_clusters=3, random_state=42))
    ]
)

# Fit pipeline on original 5-feature data
pipeline.fit(X)

# Save pipeline
joblib.dump(pipeline, 'model.joblib')

print("✅ Model pipeline saved successfully!")
```
